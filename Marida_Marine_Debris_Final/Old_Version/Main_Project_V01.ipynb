```python
import os
import rasterio
import shutil

# Function to load images using rasterio
def load_image(image_path):
    with rasterio.open(image_path) as src:
        image = src.read()
    return image

# Function to load images from a directory
def load_images_from_directory(directory):
    images = []
    for filename in os.listdir(directory):
        if filename.endswith('.tif'):
            image_path = os.path.join(directory, filename)
            image = load_image(image_path)
            images.append(image)
    return images

# Paths
train_dir = r'C:\Users\AB_\Documents\AI&Automation\Deep Learning\Final_Project\Final\data\splits\train_X.txt'
val_dir = r'C:\Users\AB_\Documents\AI&Automation\Deep Learning\Final_Project\Final\data\splits\val_X.txt'
test_dir = r'C:\Users\AB_\Documents\AI&Automation\Deep Learning\Final_Project\Final\data\splits\test_X.txt'

# Load images for training
train_images = load_images_from_directory(train_dir)
print(f"Number of training images: {len(train_images)}")

# Load images for validation
val_images = load_images_from_directory(val_dir)
print(f"Number of validation images: {len(val_images)}")

# Load images for testing
test_images = load_images_from_directory(test_dir)
print(f"Number of testing images: {len(test_images)}")
```

```python
def count_images_per_class(data_folder, labels_mapping):
    class_counts = {
        'MDebris' : 0,
        'DSargassum': 0,
        'SSargassum': 0,
        'NOMaterial': 0,
        'Ship': 0,
        'Clouds': 0,
        'MWater': 0,
        'SLWater': 0,
        'Foam': 0,
        'TWater': 0,
        'SWater': 0,
        'Waves': 0,
        'CShadows': 0,
        'Wakes': 0,
        'MixWater': 0
    }

    #iterate over each subfolder
    for folder in os.listdir(data_folder):
        folder_path = os.path.join(data_folder, folder)
        if os.path.isdir(folder_path):
            #get a list of all image files
            image_files = [file for file in os.listdir(folder_path) if file.endswith('.tif')]
        
            #iterate over each image
            for image_file in image_files:
                label = labels_mapping.get(image_file)
                if label is not None:
                    #Convert the binary labels
                    label_classes = [class_labels(i) for i in range(len(label)) if label(i) == 1]
                    for label_class in label_classes:
                        class_counts[label_class] += 1
    return class_counts

#define the class and respective 
class_labels = {
    1: 'MDebris',
    2: 'DSargassum',
    3: 'SSargassum',
    4: 'NOMaterial',
    5: 'Ship',
    6: 'Clouds',
    7: 'MWater',
    8: 'SLWater',
    9: 'Foam',
    10: 'TWater',
    11: 'SWater',
    12: 'Waves',
    13: 'CShadows',
    14: 'Wakes',
    15: 'MixWater'

}
```
